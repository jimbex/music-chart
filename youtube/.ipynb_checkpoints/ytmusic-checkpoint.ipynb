```python
with open('afrobeats-week24.txt') as file:
    read = file.readlines()
song = []
```

```python
i = 0
for x in read:
    read[i] = read[i].replace('\n', '')
    read[i] = read[i].strip()
    i += 1
```

```python
for x in read:
    if '' in read:
        read.remove('')
```

```python
read
```

```python
x = 0
while x < len(song):
    song[x] = song[x][2:-1]
    x += 1
```

```python
read = read.split('\n \n')
```

```python
for x in read:
    if x.isdigit():
        song.append(read[read.index(x) + 1])
```

```python
read
```

```python
while '' in read:
    read.remove('')
```

```python
x = 0
while x < len(read):
    read[x] = read[x].replace('\n', '')
    x +=1
```

```python
x = 0
while x < len(read):
    read[x] = read[x].split('\n')
    x += 1
```

```python
artist = []
import re
for x in read:
    if ' & ' in x:
        a = re.findall('\\n.+\\n & \\n.+\\n|\\n.*\\n & \\n.*$', x)
        
    else:
        a = re.findall('[a-z]\\n\S+ .+\\n', x)
    artist.append(a)
```

```python
artist
```

```python
x = 0
while x < len(artist):
    if len(artist[x]) > 1:
        artist[x].pop(0)
    x += 1
```

```python
x = 0
while x < len(artist):
    artist[x] = ''.join(artist[x])
    artist[x] = artist[x].replace('\n', '')
    x += 1
```

```python
data = {'Position': [x for x in range(1,101)], 'song': song, 'artist': artist}
```

```python
import pandas as pd
df = pd.DataFrame(data)
df = df.set_index('Position')
df
```

```python
df.to_csv('week21.csv')
```
