```python
import pandas as pd
df1 = pd.read_csv('album-june6.csv')
df2 = pd.read_csv('album-june7.csv')
df3 = pd.read_csv('album-june8.csv')
df4 = pd.read_csv('album-june9.csv')
df5 = pd.read_csv('album-june10.csv')
df6 = pd.read_csv('album-june11.csv')
df7 = pd.read_csv('album-june12.csv')
```

```python
albums = list(df1['album']) + list(df2['album']) + list(df3['album']) + list(df4['album']) + list(df5['album']) + list(df6['album']) + list(df7['album'])
albums = set(albums)
final = {}
```

```python
album = list(df1['title'])
for x in albums:
    if x in album:
        final[x] = list(df1['Position'][df1['title'] == x])
    else:
        final[x] = [301]
```

```python
album1 = list(df2['title'])
for x in albums:
    if x in album1:
        final[x].append(int(df2['Position'][df2['title'] == x]))
    else:
        final[x].append(301)
```

```python
album2 = list(df3['title'])
for x in albums:
    if x in album2:
        final[x].append(int(df3['Position'][df3['title'] == x]))
    else:
        final[x].append(301)
```

```python
album3 = list(df4['title'])
for x in albums:
    if x in album3:
        final[x].append(int(df4['Position'][df4['title'] == x]))
    else:
        final[x].append(301)
```

```python
album4 = list(df5['title'])
for x in albums:
    if x in album4:
        final[x].append(int(df5['Position'][df5['title'] == x]))
    else:
        final[x].append(301)
```

```python
album5 = list(df6['title'])
for x in albums:
    if x in album5:
        final[x].append(int(df6['Position'][df6['title'] == x]))
    else:
        final[x].append(301)
```

```python
album6 = list(df7['title'])
for x in albums:
    if x in album6:
        final[x].append(int(df7['Position'][df7['title'] == x]))
    else:
        final[x].append(301)
```

```python
len(final)
```

```python
albums = [x[:x.index('-') - 1 ] for x in final.keys()]
artists = [x[x.index('-') + 2:] for x in final.keys()]
post1 = [x[0] for x in final.values()]
post2 = [x[1] for x in final.values()]
post3 = [x[2] for x in final.values()]
post4 = [x[3] for x in final.values()]
post5 = [x[4] for x in final.values()]
post6 = [x[5] for x in final.values()]
post7 = [x[6] for x in final.values()]
postT = [(post1[x] + post2[x] + post3[x] + post4[x] + post5[x] + post6[x] + post7[x] )/7 for x in range(0, 134)]
```

```python
data = {'album':albums,'artist':artists,'R1':post1,'R2':post2,'R3':post3,'R4':post4,'R5':post5,'R6':post6,'R7':post7,'RT': postT}
df = pd.DataFrame(data)
```

```python
df.sort_values(by=['RT'], inplace=True)
```

```python
df['rank'] = [x for x in range(1,135)]
df = df.set_index('rank')
```
