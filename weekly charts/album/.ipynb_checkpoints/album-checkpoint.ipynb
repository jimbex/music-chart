```python
import pandas as pd
df1 = pd.read_csv('apple_album-week28.csv')
df2 = pd.read_csv('audiomack_album-week28.csv')
df3 = pd.read_csv('deezer_album-week28.csv')
df4 = pd.read_csv('boom_album-week28.csv')
```

```python
df1 = df1[['album', 'artist']]
df2 = df2[['album', 'artist']]
df3 = df3[['album', 'artist']]
```

```python
a = list(df1['album'])
a1 = a.copy()
a2 = list(df1['artist'])
x = 0
for i in a:
    if '(' in i:
        a[x] = i[: i.index('(')]
    a[x] = a[x].strip().lower()
    x += 1
```

```python
b = list(df2['album'])
b1 = b.copy()
b2 = list(df2['artist'])
x = 0
for i in b:
    if '(' in i:
        b[x] = i[: i.index('(')]
    b[x] = b[x].strip().lower()
    x += 1
```

```python
c = list(df3['album'])
c1 = c.copy()
c2 = list(df3['artist'])
x = 0
for i in c:
    if '(' in i:
        c[x] = i[: i.index('(')]
    c[x] = c[x].strip().lower()
    x += 1
```

```python
d = list(df4['album'])
d1 = d.copy()
x = 0
for i in d:
    if '(' in i:
        d[x] = i[: i.index('(')]
    d[x] = d[x].strip().lower()
    x += 1
```

```python
albums = a + b + c + d
albums = set(albums)
```

```python
albums = list(albums)
```

```python
final = {}
```

```python
x = 0
for i in albums:
    if i in a:
        final[i] = [a.index(i) + 1]
    else:
        final[i] = [500]
```

```python
x = 0
for i in albums:
    if i in b:
        final[i].append(b.index(i) + 1)
    else:
        final[i].append(510)
```

```python
x = 0
for i in albums:
    if i in c:
        final[i].append(c.index(i) + 1)
    else:
        final[i].append(450)
```

```python
x = 0
for i in albums:
    if i in d:
        final[i].append(d.index(i) + 1)
    else:
        final[i].append(520)
```

```python
x = 0
for i in albums:
    if i in a:
        final[i].append(a2[a.index(i)])
    if i in b and len(final[i]) < 6:
        final[i].append(b2[b.index(i)])
    if i in c and len(final[i]) < 6:
        final[i].append(c2[c.index(i)])
    x += 1
```

```python
x = 0
keys = list(final.keys())
for k in keys:
    if k in a:
        keys[x] = a1[a.index(k)]
    if k in b:
        keys[x] = b1[b.index(k)]
    if k in c:
        keys[x] = c1[c.index(k)]
    x += 1
```

```python
i = 0
k = list(final.keys())
for x in k:
    final[keys[i]] = final.pop(x)
    i += 1
```

```python
len(final)
```

```python
album = list(final.keys())
post1 = [x[0] for x in final.values()]
post2 = [x[1] for x in final.values()]
post3 = [x[2] for x in final.values()]
post4 = [x[3] for x in final.values()]
post5 = [x[4] if len(x) > 4 else 'NA' for x in final.values()]
postT = [(post1[x] + post2[x] + post3[x] + post4[x]/4) for x in range(0, 665)]
```

```python
data = {'album':album,'artist':post5,'apple':post1,'audiomark':post2,'deezer':post3, 'boom': post4, 'RT': postT}
df = pd.DataFrame(data)
```

```python
df.sort_values(by=['RT'], inplace=True)
```

```python
df['rank'] = [x for x in range(1,666)]
```

```python
df1 = df[['rank', 'album', 'artist']]
df1 = df1[:100]
```

```python
df1 = df1.set_index('rank')
df = df.set_index('rank')
```

```python
df2 = pd.read_csv('week27-album100.csv')
```

```python
a = list(df2['album'])
b = list(df1['album'])
c = list(df2['weeks on chart'])
i = -1
prev = []
weeks = []
for x in b:
    i += 1
    if x in a:
        prev.append(a.index(x) + 1)
        weeks.append(c[a.index(x)] + 1)
    else:
        prev.append(0)
        weeks.append(1)
df1['Previous week'] = prev
df1['weeks on chart'] = weeks
```

```python
df1.to_csv('week28-album100.csv')
df.to_csv('week28.csv')
```
