```python
import pandas as pd
df1 = pd.read_csv('apple_gospel-week24.csv')
df2 = pd.read_csv('boom_gospel-week24.csv')
df3 = pd.read_csv('youtube_gospel-week24.csv')
```

```python
df1 = df1[['song', 'artist']]
df2 = df2[['song', 'artist']]
df3 = df3[['song', 'artist']]
```

```python
a = list(df1['song'])
a1 = a.copy()
a2 = list(df1['artist'])
x = 0
for i in a:
    if '(' in i:
        a[x] = i[: i.index('(')]
    a[x] = a[x].strip().lower()
    x += 1
```

```python
b = list(df2['song'])
b1 = b.copy()
b2 = list(df2['artist'])
x = 0
for i in b:
    if 'ft' in i:
        b[x] = i.replace('ft', '(ft') + ')'
    x += 1
x = 0
for i in b:
    if '(' in i:
        b[x] = i[: i.index('(')]
    b[x] = b[x].strip().lower()
    x += 1
```

```python
c = list(df3['song'])
c1 = c.copy()
c2 = list(df3['artist'])
x = 0
for i in c:
    if '(' in i:
        c[x] = i[: i.index('(')]
    c[x] = c[x].strip().lower()
    x += 1
```

```python
songs = a + b + c
songs = set(songs)
```

```python
songs
```

```python
songs = list(songs)
```

```python
final = {}
```

```python
x = 0
for i in songs:
    if i in a:
        final[i] = [a.index(i) + 1]
    else:
        final[i] = [500]
```

```python
x = 0
for i in songs:
    if i in b:
        final[i].append(b.index(i) + 1)
    else:
        final[i].append(510)
```

```python
x = 0
for i in songs:
    if i in c:
        final[i].append(c.index(i) + 1)
    else:
        final[i].append(450)
```

```python
x = 0
for i in songs:
    if i in a:
        final[i].append(a2[a.index(i)])
    if i in b and len(final[i]) < 5:
        final[i].append(b2[b.index(i)])
    if i in c and len(final[i]) < 5:
        final[i].append(c2[c.index(i)])
    x += 1
```

```python
final
```

```python
x = 0
keys = list(final.keys())
for k in keys:
    if k in a:
        keys[x] = a1[a.index(k)]
    if k in b:
        keys[x] = b1[b.index(k)]
    if k in c:
        keys[x] = c1[c.index(k)]
    x += 1
```

```python
i = 0
k = list(final.keys())
for x in k:
    final[keys[i]] = final.pop(x)
    i += 1
```

```python
len(final)
```

```python
song = list(final.keys())
post1 = [x[0] for x in final.values()]
post2 = [x[1] for x in final.values()]
post3 = [x[2] for x in final.values()]
post4 = [x[3] for x in final.values()]
postT = [(post1[x] + post2[x] + post3[x]/3) for x in range(0, 190)]
```

```python
data = {'song':song,'artist':post4,'apple':post1,'audiomark':post2,'youtube':post3,'RT': postT}
df = pd.DataFrame(data)
```

```python
df.sort_values(by=['RT'], inplace=True)
```

```python
df['rank'] = [x for x in range(1,191)]
```

```python
df1 = df[['rank', 'song', 'artist']]
df1 = df1[:50]
```

```python
df1 = df1.set_index('rank')
df = df.set_index('rank')
```

```python
df1.to_csv('week24-hiphop50.csv')
df.to_csv('week24-hiphop.csv')
```
